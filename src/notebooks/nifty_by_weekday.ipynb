```python
from datetime import date
from dateutil.relativedelta import relativedelta
from scripts import nse
```

```python
nifty = nse.fetch_index("NIFTY 50")
```

```python
df = nifty.copy()
df['Weekday'] = df['Date'].dt.day_name()

today = date.today()
today_minus_one_year = today - relativedelta(years=1)

# Tuesdays in the last one year
filtered_df = df.query('Weekday == "Tuesday" & Date >= @today_minus_one_year')
```
